## process each dataset

```python
def get_labels_from_conll(file_path):
  with open(file_path) as file:
    file_name = os.path.basename(file_path)
    # print(file_name, file)
    lines = []
    tokens = []
    labels = []
    clean_labels = []
    entry_count = 0

    # build current entry - nope

    for line in file:
      lines.append(line)
      if line == '\n': # ignore extra lines
        continue
      line = line.split()
      tokens.append(line)
      if line[0] == '#': # ignore entry headers
        entry_count += 1
        continue
      labels.append(line[3])
      if (line[3] != 'O') and ('I-' not in line[3]): # ignore nulls and intermediates
        clean_labels.append( line[3][2:] )

    file_data = {
      'file_name': file_name,
      # 'lines': lines,
      # 'tokens': tokens,
      # 'labels': labels,
      'clean_labels': clean_labels,
      'entry_count': entry_count
    }

    return file_data
```

```python
# testing
path = '/content/drive/Shareddrives/CSCI 5980-01 F22/train_dev/en-train.conll'
file_data = get_labels_from_conll(path)

print(file_data['file_name'])
# print('entries, lines', file_data['entry_count'], len(file_data['lines']))
print('entries, clean_labels, avg', file_data['entry_count'], len(file_data['clean_labels']), len(file_data['clean_labels']) / file_data['entry_count'] )

# print(file_data['lines'][:20])
# print(file_data['tokens'][:20])
# print(file_data['labels'][:20])
print(file_data['clean_labels'][:20])

len( Counter(file_data['clean_labels']).most_common() )
# Counter(file_data['clean_labels']).most_common()
```

```python
def report_label_distribution(file_paths):
  names = []
  distributions = []
  all_labels = []
  for path in tqdm(file_paths):
    file_data = get_labels_from_conll(path)
    # print(file_data['file_name'])
    names.append( file_data['file_name'] )
    # print( Counter(file_data['clean_labels']).most_common() )
    distributions.append( Counter(file_data['clean_labels']) ) # .most_common() )
    all_labels.extend( file_data['clean_labels'] )

  file_distributions = {
    'names': names,
    'distributions': distributions,
    'all_labels': all_labels
  }

  return file_distributions
```
